<b><h1>3.The Government Health Expenditure Dataset</h1></b><br>
Lets start by looking at the Government's healthcare expenditures per year since 2006 to 2017 and the percentage of GDP it consists of.<br>
As stated above, Government Health Expenditure Dataset will be used to review the Government's increased Allocation of Money and GDP.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def sort_func1(dataset,location):
   return dataset.loc[:,location]

#Setting Dataset
df_GovHealthexpenditureDataset = pd.read_csv('government-health-expenditure.csv').set_index('financial_year')

#Sorting
df_GovHealthexpenditureDataset_oe = sort_func1(df_GovHealthexpenditureDataset,"operating_expenditure")
df_GovHealthexpenditureDataset_de = sort_func1(df_GovHealthexpenditureDataset,"development_expenditure")
df_GovHealthexpenditureDataset_pc = sort_func1(df_GovHealthexpenditureDataset,"percentage_gdp")

#Graph 1 Plotting
plt.figure(figsize=(10, 5))
plt.bar(np.arange(2006,2018,1), df_GovHealthexpenditureDataset_de,align='center',color='red', alpha=1,width=.75,label="Development Expenditure",bottom=df_GovHealthexpenditureDataset_oe)
plt.bar(np.arange(2006,2018,1), df_GovHealthexpenditureDataset_oe,align='center',alpha=1, width=.75,label="Operating Expenditure")

#Graph 1 Setting
plt.xticks(np.arange(2006,2018,1))
plt.yticks(np.arange(0,11000,1000))
bar_width = 0.6
plt.xlabel("Year")
plt.ylabel("Expenditure in Millions")
plt.title('Government Health Expenditure per year')
plt.legend(loc='best')
plt.grid()
plt.show()

#Graph 2 Plotting
fig, ax = plt.subplots()
fig.set_figwidth(10)
ax.plot(np.arange(2006,2018,1),df_GovHealthexpenditureDataset_pc)

#Graph 2 Setting
ax.set_title('Government Health Expenditure over GDP')
plt.xlabel("Year")
plt.ylabel("% of GDP")
plt.yticks(np.arange(0,6,1))
plt.xticks(np.arange(2006,2018,1))
plt.show()
```

<b><h1>5.The Number of Healthcare Specialists per 10,00 Total Population Dataset</h1></b><br>
Now that we have seen how much the government spends on Healthcare, its breakdown and the trends, we should see how this affects the amount of public healthcare specialists, namely Doctors,Nurses and Dentists.<br>

As stated above, Number of Healthcare Specialists per 10,00 Total Population Dataset, which has some missing infomation as stated above,will be used to review the amount of public healthcare specialists.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def sort_func2(dataset1,dataset2,string):
   return dataset1[dataset2 == string]

#Setting Dataset
df_HealthcareSpecialists = pd.read_csv('medical-practitioner-per-10-000-total-population.csv').set_index('year')

#Replace NA values with a -1 integer, this will be explained later.
df_HealthcareSpecialists_1 = df_HealthcareSpecialists.loc[:,'level_1']
df_HealthcareSpecialists.replace("na","-1",inplace=True)
df_HealthcareSpecialists.loc[:,'value'] = df_HealthcareSpecialists.loc[:,'value'].astype(int)

#Sorting
df_HealthcareSpecialists_dr = sort_func2(df_HealthcareSpecialists,df_HealthcareSpecialists_1,'Doctors Per 10,000 Total Population')
df_HealthcareSpecialists_de = sort_func2(df_HealthcareSpecialists,df_HealthcareSpecialists_1,'Dentists Per 10,000 Total Population')
df_HealthcareSpecialists_nur = df_HealthcareSpecialists[np.logical_and(df_HealthcareSpecialists_1 == 'Nurses Per 10,000 Total Population',df_HealthcareSpecialists.loc[:,'value'] != -1)]

#Graph 1 Plotting
plt.figure(figsize=(10, 5))
plt.plot(df_HealthcareSpecialists_dr.index, 'value', data=df_HealthcareSpecialists_dr, linewidth=1.5,label='Doctors')
plt.plot(df_HealthcareSpecialists_de.index, 'value', data=df_HealthcareSpecialists_de, linewidth=1.5,label='Dentists')
plt.plot(df_HealthcareSpecialists_nur.index, 'value', data=df_HealthcareSpecialists_nur, linewidth=1.5,label='Nurses')

#Graph 1 Setting
plt.grid()
plt.yticks(np.arange(0,90,10))
plt.xticks(np.arange(1960,2019,5))
plt.ylim(min(np.arange(0,70,5)))
plt.xlabel("Year")
plt.ylabel("Healthcare Specialists per 10,00 Total Population")
plt.title('Number of Healthcare Specialists per 10,00 Total Population')
plt.legend()
plt.show()
```

The Dengue Incidence Rate is the Number of DF/DHF local cases per 100,000 population. This is compiled yearly for every year from 2007 to 2015.<br> This dataset contains that number from every year listed.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_DengueIncidence = pd.read_csv('dengue-outbreak-statistics.csv').set_index('year')

#Dengue Incidence Rate Plotting
plt.figure(figsize=(10, 5))
plt.plot(df_DengueIncidence.index,df_DengueIncidence,label='Dengue Incidence Rate')
plt.plot(df_DengueIncidence.index,np.full(9,df_DengueIncidence.std()),label='Standard Deviation')

#Settings
plt.xticks(np.arange(2007,2016,1))
plt.yticks(np.arange(0,500,50))
plt.xlabel('Weeks')
plt.ylabel('Dengue Incidence Rate')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def sort_func6(dataset1,string1,string2,date):
   return dataset1[np.logical_and(dataset1.index == date,dataset1.loc[:,string1] == string2)]

df_denguecases = pd.read_csv('weekly-number-of-dengue-and-dengue-haemorrhagic-fever-cases.csv').set_index('year')

#Sorting for Histogram
df_denguecases_hist = df_denguecases[df_denguecases.loc[:,'type_dengue'] == 'Dengue'].loc[:,'number']

#Getting Mean for the line graph
df_denguecases_mean = df_denguecases_hist.mean(axis = 0, skipna = True);

#Sorting and Plotting for the weekly dengue cases by Year.
for number in np.arange(min(df_denguecases.index),max(df_denguecases.index)+1,1):
    plt.figure(figsize=(20, 5))
    plt.plot(sort_func6(df_denguecases,'type_dengue','Dengue',number).iloc[:,[0,2]].loc[:,'eweek'],'number',data=sort_func6(df_denguecases,'type_dengue','Dengue',number),label="Cases")
    plt.plot(np.full((54),df_denguecases_mean),label="Mean")
    plt.xticks(np.arange(0,54,1))
    plt.xlim(min(np.arange(1,50,1)))
    plt.yticks(np.arange(0,1100,100))
    plt.xlabel('Weeks')
    plt.ylabel('Number of Dengue Cases')
    plt.legend()
    plt.grid()
    plt.title('Number of Dengue Cases per week in '+number.astype(str))

#Plotting Frequency of Dengue Cases(Histogram)
plt.figure(figsize=(10, 5))
plt.hist(df_denguecases_hist,15)
plt.yticks(np.arange(0,140,20))
plt.xlabel("Number of Dengue cases")
plt.ylabel("Frequency")
plt.legend(['Cold', 'Medium', 'Hot'])
plt.grid()
plt.title('Frequency of Dengue Cases') 

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_aware = pd.read_csv('omnibus-survey-2009-2011.csv').set_index('year')

color = ['lightblue','blue','darkblue','indianred','red','maroon']
df_aware_gender_overall = df_aware[df_aware.loc[:,"gender"] == "Overall"];
df_aware_gender_2009 = df_aware[np.logical_and(df_aware.loc[:,"age_group"] != "Overall",df_aware.index == 2009)]
df_aware_gender_2011 = df_aware[np.logical_and(df_aware.loc[:,"age_group"] != "Overall",df_aware.index == 2011)]

plt.figure(figsize=(10, 5))
for label in np.arange(0,df_aware_gender_2011.loc[:,"knowledge_on_wholegrain"].to_numpy().size,1):
    plt.scatter(df_aware_gender_2009.loc[:,"knowledge_on_wholegrain"].to_numpy()[label]*100,
                df_aware_gender_2009.loc[:,"aware_of_transfat"].to_numpy()[label]*100,
                label=df_aware_gender_2009.loc[:,"age_group"].to_numpy()[label]+" "+df_aware_gender_2009.loc[:,"gender"].to_numpy()[label],
                c=color[label])
plt.xlabel("Percentage Knew about Wholegrain")
plt.ylabel("Percentage aware about Trans-fat")
plt.title("Healthcare Survey 2009")
plt.legend(loc="best")
plt.yticks(np.arange(0,105,5))
plt.xticks(np.arange(0,105,5))
plt.grid()

plt.figure(figsize=(10, 5))
for label in np.arange(0,df_aware_gender_2011.loc[:,"knowledge_on_wholegrain"].to_numpy().size,1):
    plt.scatter(df_aware_gender_2011.loc[:,"knowledge_on_wholegrain"].to_numpy()[label]*100,
                df_aware_gender_2011.loc[:,"aware_of_transfat"].to_numpy()[label]*100,
                label=df_aware_gender_2011.loc[:,"age_group"].to_numpy()[label]+" "+df_aware_gender_2011.loc[:,"gender"].to_numpy()[label],
                c=color[label])
plt.xlabel("Percentage Knew about Wholegrain")
plt.ylabel("Percentage aware about Trans-fat")
plt.title("Healthcare Survey 2011")
plt.legend(loc="best")
plt.yticks(np.arange(0,105,5))
plt.xticks(np.arange(0,105,5))
plt.grid()  

plt.figure(figsize=(10, 5))
for label in np.arange(0,df_aware_gender_overall.loc[:,"knowledge_on_wholegrain"].to_numpy().size,1):
    plt.scatter(df_aware_gender_overall.loc[:,"knowledge_on_wholegrain"].to_numpy()[label]*100,
                df_aware_gender_overall.loc[:,"aware_of_transfat"].to_numpy()[label]*100,
                label=df_aware_gender_overall.loc[:,"gender"].to_numpy()[label]+" "+str(df_aware_gender_overall.index[label]),
                c=color[label])
plt.xlabel("Percentage Knew about Wholegrain")
plt.ylabel("Percentage aware about Trans-fat")
plt.title("Average Overall")
plt.legend(loc="best")
plt.yticks(np.arange(0,105,5))
plt.xticks(np.arange(0,105,5))
plt.grid() 
```
